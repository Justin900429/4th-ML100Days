- 作業目標 : 用 iris (dataset.load_iris()) 資料嘗試跑 hierarchical clustering

```python
from sklearn import datasets
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
%matplotlib inline

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
data = pd.DataFrame(X, columns=[i.replace(' (cm)', "") for i in iris.feature_names])
```

```python
data.head()
```

載入 相關套件 並 執行 hierarchical clustering 實驗 ...

```python
estimators = [('hc_word', AgglomerativeClustering(n_clusters=3, linkage="ward")),
              ('hc_complete', AgglomerativeClustering(n_clusters=3, linkage='complete')),
              ('hc_average', AgglomerativeClustering(n_clusters=3, linkage='average'))]
```

```python
# Original graph
sns.scatterplot(x='sepal length', y='sepal width', data=data, hue=y)
```

```python
for name, clf in estimators:
    print("Using {} training".format(name))
    clf.fit(X)
    sns.scatterplot(x='sepal length', y='sepal width', data=data, hue=clf.labels_)
    plt.show()
    
```
