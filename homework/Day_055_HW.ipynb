- 作業目標 : iris (dataset.load_iris()) 資料嘗試跑 kmeans (可以測試不同的群數 , init 等)

```python
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

### 載入 相關套件 並 執行 kmean 實驗 ...

- 實驗測試不同的群數
- 實驗測試不同的初始值
- 呈現結果

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.DataFrame(X, columns=[i.replace(" (cm)", "") for i in iris.feature_names])
target = pd.DataFrame(y, columns=['target'])
```

```python
iris_data = pd.concat([data, target], axis=1)
```

```python
iris_data.head()
```

```python
sns.scatterplot(x='sepal length', y='sepal width', data=iris_data, hue='target')
```

```python
from sklearn.cluster import KMeans

numbers = range(3, 7)
plt.figure(figsize=(15, 15))

for number in numbers:
    plt.subplot(2, 2, number - 2)
    estimator = KMeans(n_clusters=number, random_state=0)
    estimator.fit(data)
                
    sns.scatterplot(x='sepal length', y='sepal width', data=iris_data, hue=estimator.labels_)
```

```python
from sklearn.cluster import KMeans

numbers = range(3, 7)
plt.figure(figsize=(15, 15))

for number in numbers:
    plt.subplot(2, 2, number - 2)
    estimator = KMeans(n_clusters=3, random_state=number)
    estimator.fit(data)
                
    sns.scatterplot(x='sepal length', y='sepal width', data=iris_data, hue=estimator.labels_)
```
