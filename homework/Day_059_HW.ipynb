### 用 digits (3個數字即可 , dataset.load_digits(n_class=3)) 資料嘗試跑 PCA


```python
from sklearn import datasets

digits = datasets.load_digits(n_class=3)
X = digits.data
y = digits.target
```

載入 相關套件 並 執行 PCA 實驗 ...

```python
import matplotlib.pyplot as plt
from math import sqrt

length = 10

plt.figure(1, figsize=(12, 12))

for i in range(length):
    plt.subplot(5, length - 5, i + 1)
    plt.imshow(X[i].reshape(8, 8), cmap=plt.cm.gray_r, interpolation='nearest')


plt.show()
```

```python
from sklearn import decomposition   

X = decomposition.PCA(n_components=36).fit_transform(X)
```

```python
plt.figure(1, figsize=(12, 12))

for i in range(length):
    plt.subplot(5, length - 5, i + 1)
    plt.imshow(X[i].reshape(6, 6), cmap=plt.cm.gray_r, interpolation='nearest')
    
plt.show()
```
