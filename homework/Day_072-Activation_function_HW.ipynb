# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline


def relu(x):
    return max(0, x)

x_range = np.linspace(-10, 10, 100)
y = [relu(number) for number in x_range]

plt.title("ReLU")
plt.plot(x_range, y, 'k')
```

```python
def first_deriv_relu(number):
    return 0 if number <= 0 else 1

y = [first_deriv_relu(number) for number in x_range]

plt.title("First order derivative relu")
plt.plot(x_range, y, 'y')
```
